#### 불러온 로그데이터에 대한 정보 입니다.

* ```eventcategory```: 앱의 이벤트에 대한 카테고리입니다. 크게 다음의 종류가 있습니다.
  - ```install```: 앱설치
  - ```launch```: 앱실행 
  - ```deeplinkLaunch```: 딥링크 통한 앱실행
  - ```goal```: 앱 내에서의 활동 여부
  - ```exit```: 앱 종료
  - ```foreground```, background```: 앱을 foreground, background로 전환
  - ```launchlnSession```: Session의 시작.
* ```isfirstactivity```: 해당 이벤트가 해당 유저에 한하여 첫 번째인지 혹은 두 번째 이상인지를 알려주는 값입니다. 
  - boolean 형식으로, 해당 이벤트가 해당 유저에 한하여 첫 번째 이벤트이면 True, 아니면 False를 갖습니다.
* ```apppackagename```: 앱의 패키지의 고유한 이름입니다. 안드로이드의 경우 applicationId을, iOS의 경우 Bundle ID를 가져오게 됩니다.
* ```appversion```: kmong 어플리케이션의 버전입니다.
* ```devicetype```: 사용자가 kmong app을 실행하고 있는 기기의 제품명입니다.
* ```devicemanufacturer```: 사용자가 kmong app을 실행하고 있는 기기의 제조회사입니다. 
* ```osversion```: 사용자가 kmong app을 실행하고 있는 기기의 os 버전입니다.
* ```canonicaldeviceuuid```: 사용자가 kmong app을 실행하고 있는 기기의 고유 식별자입니다. (사용자를 구분하는 식별자로 이용할 수 있습니다)
* ```sourcetype```: 해당 로그를 만든 고객이 유입된 경로입니다. 유료 광고 채널, 바이럴 채널, 앱 마켓 등으로 구분합니다.
* ```channel```: sourcetype과 동일하지만 조금 더 세분화되어 있습니다.
  - unattributed, WEB, google-play, m_naver, google, (not set), google.adwords, m_naverpowercontents, pc_naver, apple.searchads, facebook, m_daum
* ```params_campaign```: 마케터가 입력한 캠페인 파라미터 중 캠페인명입니다. 
* ```params_medium```: 마케터가 입력한 캠페인 파라미터 중 매체입니다.
* ```params_term```: 마케터가 입력한 캠페인 파라미터 중 키워드입니다.
* ```inappeventcategory```: 인앱이벤트 식별자값으로 인앱이벤트를 지칭하는 3개의 값(category > action > label) 중 가장 상위에 속하는 구분값입니다.
  - ```eventcategory```가 goal인 경우에만 제공됩니다.
  - 차후에 읽어올 funnel 데이터셋과 연결할 수 있습니다.
  - ex) seller_selling_history.view, gig_detail.view
* ```inappeventlabel```: categoryid와 동일합니다.
  - 차후에 읽어올 category 데이터셋과 연결할 수 있습니다.
  - 인앱이벤트를 지칭하는 3개의 값(category > action > label) 중 마지막에 속하는 구분값입니다.
* ```eventdatetime```: 모바일 클라이언트 상에서의 실제 이벤트 발생시간입니다. ISO 8601 표준에 따라서 제공됩니다. 
* ```isfirstgoalactivity```: goal 이벤트의 경우, 해당 Goal Label, Description, Key, Category를 가진 이벤트가 해당 유저에 한하여 첫 번째인지 혹은 두 번째 이상 인지를 알려주는 boolean 값입니다. 예를 들어 동일한 유저로부터 두 번 이상의 같은 Goal 이벤트가 발생한 경우(e.g. 동일한 물건 재구매 등) 이 값은 false가 됩니다.

#### canonicaldeviceuuid 컬럼명을 userid로 변경

````python
# user_id로 새로운 컬럼 생성
# canonicaldeviceuuid컬럼은 삭제해도 되지만 원본 데이터를 삭제하는것은 문제가 발생될수 있으므로
# 뒷쪽에서 log데이터를 모두 정리한뒤 필요한 컬럼만 뽑아 새로운 변수에 할당하는것이 적절함
# 추가적으로 아래에 정리될 컬럼들도 같은 이유로 삭제 없이 새로운 컬럼 생성후 최종적으로 정리 및 컬럼명 변경함

# canonicaldeviceuuid컬럼의 값을 새로운 user_id 컬럼에 복사합니다(넣어줍니다).
log["userid"] = log["canonicaldeviceuuid"]

# log 변수에 할당된 데이터의 행렬 사이즈를 출력합니다 결과는 (row, column) 으로 표시됩니다.
print(log.shape)

# log 데이터 중, canonicaldeviceuuid와 userid 두 컬럼의 상위 5개만 출력하여 값이 잘들어 갔는지 확인힙니다
# 편의상 head() 를 이용하여 상위 5개만 확인합니다. 
log[["canonicaldeviceuuid", "userid"]].head()
````

#### osversion 컬럼 정보를 ostype 와 osversion으로 나누기

````python
# ostype를 찾아주는 함수를 정의 합니다.
def find_ostype(osversion):
    # osversion에 iOS라는 값이 있으면
    if "iOS" in osversion:
        # 아이폰이라고 간주하여 iOS라는 값을 반환합니다
        return "iOS"
    # 그렇치 않으면
    else:
        # 안드로이드라고 간주하고 Android라는 값을 반환합니다.
        return "Android"    

# osversion를 찾아주는 함수를 정의합니다.
def find_osversion(osversion):
    # osversion에 iOS와 Android라는 값이 있으면 이를 제거합니다 repalce()라는 메소드는 특정 값을 대체해주는 기능을 합니다.
    # 결과적으로 osversion에 iOS와 Android라는 명이 제거된 버전(숫자)만 남게 되어 출력됩니다.
    return osversion.replace("iOS", "").replace("Android", "")  

# 구현된 함수를 이용하여 apply 메소드를 통해 ostype(clean) / osversion(clean) 컬럼을 생성합니다.
log["ostype(clean)"] = log["osversion"].apply(find_ostype)
log["osversion(clean)"] = log["osversion"].apply(find_osversion)

# 생성된 ostype(clean)",osversion(clean) 과 osversion 동시에 출력하여 값이 잘들어 갔는지 확인합니다. 상위 5개 값만 보여집니다.
log[["osversion", "ostype(clean)","osversion(clean)"]].head()
````

#### devicemanufacturer 컬럼 정리

````python
# devicemanufacturer의 컬럼값을 vales_counts() 함수로 확인해 보니 삼성/애플/LG 가 전체 데이터의 98.2%정도 이므로 나머지 제조사는
# 분석의 편의를 위해 Others 값으로 정리 하고 삼성/애플/LG제조사는 비슷한 이름으로 나눠져 있을 경우를 생각하여
# 제조사를 찾아주는 함수를 정의 한뒤 정리합니다.

# 제조사를 찾아주는 함수 정의 합니다.
def find_devicemanufacturer(devicemanufacturer):
    # devicemanufacturer 값에 samsung이라는 값이 들어가 있으면
    if "samsung" in devicemanufacturer:
        # s를 대문자(S)로 바꿔서 Samsung이라는 값을 반환해줍니다.
        return "Samsung"
    # devicemanufacturer 값에 LGE라는 값이 있거나 LG Electronics 값이 있으면
    elif "LGE" in devicemanufacturer or "LG Electronics" in devicemanufacturer:
        # 모두 동일하게 LG라는 값을 반환해줍니다.
        return "LG"
    # devicemanufacturer 값에 Apple이라는 값이 들어가 있으면
    elif "Apple" in devicemanufacturer:
        # Apple이라는 값을 반환해줍니다
        return "Apple"
    # 위의 조건이 모두 아니면
    else:
        # Other라는 값을 반환해줍니다
        return "Others"

# devicemanufacturer_simple의 새로운 컬럼에 apply() 메소드를 이용하여 값을 넣어줍니다.
log["devicemanufacturer_simple"] = log["devicemanufacturer"].apply(find_devicemanufacturer)

# value_counts() 함수를 통해 값의 종류별 갯수를 확인합니다.
log["devicemanufacturer_simple"].value_counts()
````

#### channel 컬럼 정리

````python
# channel 컬럼을 value_counts() 로 확인하여 내부 값을 보면 같은 같은 싸이트 인데 값의 이름이 다양하여 정리가 필요합니다.
# 특정 싸이트 명이 있는 값들은 하나의 값으로 바꾸줍니다.단, 데이터의 빈값들이 있으므로 그것들을 감안하여 코딩을 해야합니다.

# 채널값들을 찾아주는 find_channel 함수를 정의 합니다.
def find_channel(channel):
    # 우선먼저 channel 값에 비어있는 값(NaN)이 들어가 있으면
    if pd.isna(channel) == True:
        # 아무런 처리도 하지 않고 그대로 들어온 값을 반환해 줍니다.
        return channel
    # channel에 WEB이라는 값이 들어가 있으면
    elif "WEB" in channel:
        # web로 반환해 줍니다.
        return "web"
    # channel에 google이라는 값이 들어가 있으면
    elif "google" in channel:
        # google를 반환해 줍니다.
        return "google"
    # channel에 naver이라는 값이 들어가 있으면
    elif "naver" in channel:
        # naver를 반환해 줍니다
        return "naver"
    # channel에 daum이라는 값이 들어가 있으면
    elif "daum" in channel:
        # daum을 반환해 줍니다.
        return "daum"
    # channel에 apple이라는 값이 들어가 있으면
    elif "apple" in channel:
        # apple을 반환해 줍니다.
        return "apple"
    # 위조건이 모두 해당되지 않으면
    else:
        # 들어온 값을 그대로 반환해 줍니다.
        return channel

# apply()를 이용하여 새로운 channel(clean) 컬럼을 생성해 줍니다.
log["channel(clean)"] = log["channel"].apply(find_channel)
# 새로 생성된 channel(clean)의 값을 value_counts() 메소드를 이용하여 종류별 갯수를 확인힙니다
log["channel(clean)"].value_counts()
````

#### inappeventcategory 컬럼 viewcategory / viewid / viewaction 으로 분류하여 정리

````python
# funnel 데이터와 log데이터를 합칠때 사용하는 inappeventcategory를 정리합니다.
# inappeventcategory 컬럼의 값들을 "_"(언더바) 와 "."(점) 을 기준으로 분류하여 viewcategory / viewid / viewaction 컬럼으로 나누어줍니다.
# viewcategory - inappeventcategory에서 언더바(_)를 기준으로 왼쪽 텍스트만 가져옵니다. 언더바가 없을 경우 점(.)의 왼쪽 텍스트만 가져옵니다.
# vewid - inappeventcategory에서 점(.)의 왼쪽 텍스트만 가져옵니다.
# viewaction - inappeventcategory에서 점(.)의 오른쪽 텍스트만 가져옵니다.


# viewcategory컬럼의 값을 찾아줄 find_viewcategory 함수를 정의 합니다.
def find_viewcategory(inappeventcategory):
    # inappeventcategory에 빈값이 들어 온다면
    if pd.isnull(inappeventcategory):
        # inappeventcategory값 그대로를 반환합니다.
        return inappeventcategory
    # 그러치 않으면
    else:
        # inappeventcategory에서 점(.)의 왼쪽을 가져와 bridge라는 이름의 변수에 할당합니다.
        bridge = inappeventcategory.split(".")[0]
        # 그리고 bridge값의 왼쪽에서 다시 한 번 언더바(_)의 왼쪽을 가져와 viewcategory변수에 할당합니다.
        viewcategory = bridge.split("_")[0]
        # 최종적으로 viewcategory를 반환합니다. 
        return viewcategory

# viewid컬럼 값을 찾아줄 find_viewid 함수를 정의 합니다.
def find_viewid(inappeventcategory):
    # inappeventcategory에 빈값이 들어 온다면
    if pd.isnull(inappeventcategory):
         # inappeventcategory값 그대로를 반환합니다.
        return inappeventcategory
    # 그러치 않으면
    else:
        # inappeventcategory에서 점(.)의 왼쪽을 가져와 viewid이라는 이름의 변수에 할당합니다.
        viewid = inappeventcategory.split(".")[0]
        # 최종적으로 viewid를 반환합니다.
        return viewid

# viewaction컬럼 값을 찾아줄 find_viewaction 함수를 정의 합니다.
def find_viewaction(inappeventcategory):
    # inappeventcategory에 빈값이 들어 온다면
    if pd.isnull(inappeventcategory):
        # inappeventcategory값 그대로를 반환합니다.
        return inappeventcategory
    # 그러치 않으면
    else:
        # inappeventcategory에서 점(.)의 오른쪽을 가져와 viewaction이라는 이름의 변수에 할당합니다.
        viewaction = inappeventcategory.split(".")[1]
        # 최종적으로 viewaction을 반환합니다.
        return viewaction

# 위 세 개의 함수(find_viewcategory, find_viewid, find_viewaction)를 apply() 메소드를 이용하여 
# viewcategory, viewid, viewaction 라는 새로운 컬럼을 생성합니다.
log["viewcategory"] = log["inappeventcategory"].apply(find_viewcategory)
log["viewid"] = log["inappeventcategory"].apply(find_viewid)
log["viewaction"] = log["inappeventcategory"].apply(find_viewaction)

# 새로생성된 컬럼과 inappeventcategory을 출력하여 확인합니다.
log[["inappeventcategory","viewcategory", "viewid", "viewaction"]].head()
````

광고비를 사용해야 합니다. 따라서 광고채널별 유입량을 비교해보고 어디에 집중해야 할지를 판단하는것은 중요하다고 생각합니다.  이번에는 위에서 정리된 데이터를 바탕으로 분석 및 시각화를 진행해 보겠습니다.

### 1) pageview 데이터 생성 및 시각화
 - 정리한 로그 데이터를 바탕으로, 로그가 기록된 2018년 09월 28일부터 29일까지의 48시간동안 발생한 페이지뷰의 변화량을 출력하겠습니다
 - 사용자가 서비스를 한 번 이용할 때 마다 페이지뷰가 한 번 올라갔다고 가정하겠습니다.

````python
# data의 "eventdatetime" 컬럼의 시간을 우선 빠른순으로 정렬한뒤 경과된 시간측정을 위한 컬럼을 생성하고
# 새로 생성된 "경과시간"컬럼을 그래프로 그리기 위한 단위를 나누는 작업을 진행 합니다.
# 또한 rowuuid를 통해 페이지뷰를 셀수 있는 컬럼을 새로 생성합니다.

# data값의 인덱스('userid','rowuuid')를 reset_index()를 통해 컬럼으로 뽑아낸 뒤 그값을 page_data 변수에 넣어줍니다.
page_view_data = data.reset_index()

# page_data를 eventdatetime컬럼을 기준으로 sort_values()를 이용하여 시간순(시간이 빠른)으로 정렬한뒤 다시 page_data 변수에 넣어줍니다.
page_view_data = page_view_data.sort_values(by = "eventdatetime")

# 정렬된 page_data를 다시 reset_index()를 통해 index값을 컬럼으로 뽑아낸뒤 그값을 drop()를 통해 삭제합니다. 
# 그리고 다시 page_data 변수에 넣어줍니다.
page_view_data = page_view_data.reset_index().drop("index", axis = "columns")

# 경과시간을 파악하기 위해 eventdatetime 컬럼의 첫번째 시간(값)을 기준으로 eventdatetime 모든값들을 수정합니다.
# 각각의 eventdatetime값 - eventdatetime의 첫번째 값을 하여 구해진 경과시간 값을 elasped라는 새로운 컬럼에 넣어줍니다.
page_view_data["elasped"] = page_view_data["eventdatetime"] - page_view_data.loc[0,"eventdatetime"]

# elasped컬럼의 값들을 "일"별로 뽑아 time_day 라는 변수에 넣어줍니다.
time_day = page_view_data["elasped"].dt.components.days
# elasped컬럼의 값들을 "시간"별로 뽑아 time_hour 라는 변수에 넣어줍니다.
time_hour = page_view_data["elasped"].dt.components.hours
# elasped컬럼의 값들을 "분"별로 뽑아 time_minute 라는 변수에 넣어줍니다.
time_minute = page_view_data["elasped"].dt.components.minutes
# elasped컬럼의 값들을 "초"별로 뽑아 time_second 라는 변수에 넣어줍니다.
time_second = page_view_data["elasped"].dt.components.seconds

# "일","시간","분","초" 의 값을 모두 분으로 환산하여 page_data의 elasped컬럼에 다시 넣어줍니다.
page_view_data["elasped"] = time_day*24*60 + time_hour*60 + time_minute + time_second/60

# elasped컬럼의 값을 astype(int)을 이용하여 int형태로 변형한뒤 새로운 elasped(int)컬럼에 넣어줍니다.
page_view_data["elasped(int)"] = page_view_data["elasped"].astype(int)

# 새로 생성된 elasped(int)값을 60단위(1시간)로 그룹화하여 다시 elasped(int)컬럼에 넣어줍니다.
page_view_data["elasped(int)"] = (page_view_data["elasped(int)"] // 60) * 60

# 페이지뷰를 카운터 하기 위해 rowuuid컬럼에 notnull()을 이용하여 새로운 "count" 컬럼에 넣어줍니다.
page_view_data["count"] = page_view_data["rowuuid"].notnull()
````

## <center> *3. 크몽기업의 로그데이터를 통한 비효율 상품카테고리 분석*
기업의 매출을 증대시키기위해 서비스로 고객을 유입시키는것 다음으로 중요한것은 유입된 고객이 실제 제품이나 서비스를 구매하게 하는것이라 생각합니다.    따라서 이번에는 들어온 고객들이 많이 찾는 상품 카테고리에서 실제 구매가 잘 이루어지지 않는 카테고리를 찾아내고 해결방안을 제시해보겠습니다.

### 1) 유저들의 상품검색 대비 실제 결제율이 저조한 카테고리 탐색
funnel데이터에는 고객이 서비스에 들어 와서 결제를 진행하기까지 활동한 다양한 기록이 있습니다. 특히 이번에 사용할 컬럼은 viewid desc컬럼으로 고객이 무슨 상품을 찾아서 보고 있는지가 나와있으며 실제 결제를 진행했는지 여부가 나와있습니다. 이것을 통해 고객이 어떠한 상품을 많이 찾고 있으며 결제는 어떻게 이루어지고 있는지 비율로 확인해보겠습니다.

````python
# data의 viewid desc컬럼에서 "상품상세" 와 "상품상세-상품선택"이 있는 데이터만 가져와 searching_data변수에 넣어줍니다.
searching_data = data.loc[(data["viewid desc"] == "상품상세") | (data["viewid desc"] == "상품상세-상품선택")]
# data의 viewid desc컬럼에서 "결제완료"가 있는 데이터만 가져와 paying_data변수에 넣어줍니다.
paying_data = data.loc[(data["viewid desc"] == "결제완료")]

# searching_data의 categoryname컬럼에 들어 있는 값을 value_counts를 이용하여 데이터별 수량을뽑고 reset_index를 이용하여 인덱스를
# 초기화 합니다. 그 후 index로 되어있는 컬럼은 category_name으로 categoryname로 되어있는 컬럼은 seaching_counts로 이름을 변경합니다.
searching_data_user = (searching_data["categoryname"]
                       .value_counts()
                       .reset_index()
                       .rename(columns = {"index" : "category_name", "categoryname" : "searching_counts"})
                      )

# paying_data의 categoryname컬럼에 들어 있는 값을 value_counts를 이용하여 데이터별 수량을뽑고 reset_index를 이용하여 인덱스를
# 초기화 합니다. 그 후 index로 되어있는 컬럼은 category_name으로 categoryname로 되어있는 컬럼은 paying_counts로 이름을 변경합니다.
paying_data_user = (paying_data["categoryname"]
                    .value_counts()
                    .reset_index()
                    .rename(columns = {"index" : "category_name", "categoryname" : "paying_counts"})
                   )

# searching_data 와 paying_data를 merge를 이용하여 합친후 searchin_n_paying_user변수에 넣어줍니다.
# how옵션을 이용하여 inner(서로없는것 제외) merge를 합니다.
searchin_n_paying_user = searching_data_user.merge(paying_data_user, how = "inner")
# searchin_n_paying_user의 paying_counts컬럼을 seraching_counts컬럼으로 나누준값에 100을 곱하여 새로운 paying_ratio컬럼에 넣어줍니다.
searchin_n_paying_user["paying_ratio"] = (searchin_n_paying_user["paying_counts"]/searchin_n_paying_user["searching_counts"])*100

# searchin_n_paying_user의 searching_counts값이 300이상인 값만 추린뒤 sort_value를 이용하여 paying_ratio를 기준으로 큰값순으로 정렬합니다.
# 인덱스를 초기화한뒤 컬럼으로 나온index는 제거하고 다시 searchin_n_paying_user변수에 넣어줍니다.
searchin_n_paying_user = (searchin_n_paying_user
                          .loc[searchin_n_paying_user["searching_counts"] >=300]
                          .sort_values(by = "paying_ratio", ascending = False)
                          .reset_index(drop = True)
                         )

#그래프의 크기는 (10,18)로 정해줍니다.
plt.figure(figsize = (10, 18))
# seaborn의 barplot을 이용하여 searchin_n_paying_user데이터를 그려줍니다. y값은 category_name x값은 paying_ratio로 하여 orient옵션은
# h로 하여 세로형태의 그래프로 나타냅니다.
sns.barplot(data = searchin_n_paying_user, y = "category_name", x = "paying_ratio", orient = "h")

# 판매율이 저조한 카테고리를 따로 추려 보여주기 위한 표를 작성해줍니다.
# searchin_n_paying_user에서 paying_ratio값이 0.5이하인 값만 추리고 sort_value를 이용하여 searching_counts를 기준으로 큰순으로 정렬합니다.
# 인덱스는 초기화 하고 컬럼으로 나온 index는 제거해준뒤 low_paying_ratio_category 변수에 넣어줍니다.
low_paying_ratio_category = (searchin_n_paying_user
                             .loc[searchin_n_paying_user["paying_ratio"] <= 0.5]
                             .sort_values(by = "searching_counts",ascending = False)
                             .reset_index(drop = True)
                            )
# low_paying_ratio_category의 상위 5개 값을 확인합니다.
low_paying_ratio_category.head()
````
